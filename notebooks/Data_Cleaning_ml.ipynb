```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Cleaned_Hotel_Reservations.csv')
```

```python
df.dtypes
```

```python
df['arrival_date_full'] = pd.to_datetime(
    df['arrival_date_full'],
    format='%d-%m-%Y'
)
```

```python
df['arrival_date_full'].isnull()
```

```python
df = df.drop(columns=['Booking_ID'])
```

```python
Q1 = df['avg_price_per_room'].quantile(0.25)
Q3 = df['avg_price_per_room'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[
    (df['avg_price_per_room'] >= lower) &
    (df['avg_price_per_room'] <= upper)
]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_cols = df.select_dtypes(include="object").columns

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
print("Final shape:", df.shape)
df.to_csv('Cleaned_Hotel_Reservations.csv', index=False)
```

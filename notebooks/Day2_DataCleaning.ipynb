```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Hotel Reservations.csv')
```

```python
print("Colums with zero values and their counts:")
for col in df.columns:
    zero_count=(df[col] ==0).sum()
    if zero_count > 0:
        print(f"Column'{col}:{zero_count} zeroes")
```

```python
#zero avg price per room investigation
zero_price = df[df['avg_price_per_room'] == 0]
print(f"Zero-priced bookings: {len(zero_price)}")
print("\nMarket Segment breakdown:")
display(zero_price['market_segment_type'].value_counts())
print("\nBooking Status breakdown:")
display(zero_price['booking_status'].value_counts())
print("\nRoom Type breakdown:")
display(zero_price['room_type_reserved'].value_counts())
```

```python
#zero no of adults investigation
zero_adults=df[df['no_of_adults']==0]
print(f"Bookings with 0 adults: {len(zero_adults)}")
display(zero_adults['booking_status'].value_counts())
```

```python
#children outlier investigation
print("no_of_children distribution:")
display(df['no_of_children'].value_counts().sort_index())
```

```python
zero_nights = df[(df['no_of_weekend_nights'] == 0) & (df['no_of_week_nights'] == 0)]
print(f"Bookings with 0 total nights: {len(zero_nights)}")
display(zero_nights['booking_status'].value_counts())
```

```python
print(f"Bookings with lead_time > 365 days: {(df['lead_time'] > 365).sum()}")
print(df['lead_time'].describe())
```

```python
# Identify object-type columns and display unique values
object_cols = df.select_dtypes(include=['object']).columns

print(f"Categorical columns (object type): {list(object_cols)}\n")

for col in object_cols:
    unique_vals = df[col].unique()
    print(f"Unique values in '{col}' ({len(unique_vals)} values):")
    print(unique_vals)
    print("-" * 30)
```

```python
#Identify integer type and display unique values
int_cols=df.select_dtypes(include=['int64']).columns

print(f"Categorical columns (int type):{list(int_cols)}")
for col in int_cols:
  unique_vals=df[col].unique()
  print(f"Unique values in '{col}' ({len(unique_vals)} values):")
  print(unique_vals)
  print("-"*30)
```

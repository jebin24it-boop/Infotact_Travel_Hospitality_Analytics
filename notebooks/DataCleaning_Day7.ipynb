```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Hotel_Reservations_2.csv')
```

```python
dupe_check_cols = [c for c in df.columns if c != 'Booking_ID']
dupes = df[df.duplicated(subset=dupe_check_cols, keep=False)]
print(dupes.sort_values(by=dupe_check_cols).head(10))
print("Number of duplicate groups:", df.duplicated(subset=dupe_check_cols).sum())
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Hotel Reservations.csv')
```

```python
df['booking_status']
```

```python
df['booking_status'].astype(str)
```

```python
#Converting the status of booking to binary values and Canceled :0 and Not canceled as 1
status_mapping = {'Not_Canceled': 1, 'Canceled': 0}
df['booking_status'] = df['booking_status'].replace(status_mapping)
print("Booking status column converted to binary.")
```

```python
print(df['booking_status'].value_counts())
print("Booking status column verified.")
print(df['booking_status'])
```

```python
print(df['booking_status'].isnull().sum())
```

```python
df['market_segment_type'].unique()
```

```python
nan_columns = df.isnull().sum()
nan_columns = nan_columns[nan_columns > 0]

if not nan_columns.empty:
    print("Columns with NaN values and their counts:")
    print(nan_columns)
else:
    print("No columns with NaN values found in the DataFrame.")
```

```python
# Identify object-type columns and display unique values
object_cols = df.select_dtypes(include=['object']).columns

print(f"Categorical columns (object type): {list(object_cols)}\n")

for col in object_cols:
    unique_vals = df[col].unique()
    print(f"Unique values in '{col}' ({len(unique_vals)} values):")
    print(unique_vals)
    print("-" * 30)
```

```python
import pandas as pd
import numpy as np
```

```python
# Load the partially cleaned dataset
df = pd.read_csv('Hotel_Reservations_1.csv')
display(df.head())
```

```python
# Drop the 'Unnamed: 0' column if it exists and is redundant
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
    print("Dropped 'Unnamed: 0' column.")
else:
    print("'Unnamed: 0' column not found or already dropped.")
```

```python
initial_rows = df.shape[0]
invalid_date_rows_to_drop = df[df['arrival_date_full'] == '29-02-2018'].shape[0]
df = df[df['arrival_date_full'] != '29-02-2018']
print(f"Dropped {invalid_date_rows_to_drop} rows with invalid date '29-02-2018'.")
print(f"Remaining rows: {df.shape[0]}")
```

```python
zero_nights_rows_to_drop = df[df['total_nights'] == 0].shape[0]
df = df[df['total_nights'] != 0]
print(f"Dropped {zero_nights_rows_to_drop} rows with zero total nights.")
print(f"Remaining rows: {df.shape[0]}")
```

```python
children_outlier_rows_to_drop = df[df['no_of_children'] > 3].shape[0]
df = df[df['no_of_children'] <= 3]
print(f"Dropped {children_outlier_rows_to_drop} rows with more than 3 children.")
print(f"Remaining rows: {df.shape[0]}")
```

```python
print('Final DataFrame shape:', df.shape)
print('\nFinal DataFrame Info:')
display(df.info())
```

```python
condition = (df['avg_price_per_room'] == 0) & ((df['total_guests'] == 0) | ((df['no_of_adults'] == 0) & (df['no_of_children'] == 0)))
filtered_df = df[condition]

print(f"Number of rows with avg_price_per_room = 0 AND (total_guests = 0 OR (no_of_adults = 0 AND no_of_children = 0)): {filtered_df.shape[0]}")
display(filtered_df.head())
```

```python
condition_2 = (df['avg_price_per_room'] == 0) & ((df['no_of_adults'] > 0) | (df['no_of_children'] > 0))
filtered_df_2 = df[condition_2]

print(f"Number of rows with avg_price_per_room = 0 AND (no_of_adults > 0 OR no_of_children > 0): {filtered_df_2.shape[0]}")
display(filtered_df_2.head())
```

```python
rows_to_drop_count = filtered_df_2.shape[0]
df = df[~condition_2]

print(f"Dropped {rows_to_drop_count} rows where avg_price_per_room was 0 but guests were present.")
print(f"Remaining rows in DataFrame: {df.shape[0]}")
```

```python
rows_to_drop_count = filtered_df_2.shape[0]
df = df[~condition_2]

print(f"Dropped {rows_to_drop_count} rows where avg_price_per_room was 0 but guests were present.")
print(f"Remaining rows in DataFrame: {df.shape[0]}")
```

```python
df['required_car_parking_space'].unique()
```

```python
df = df.drop(columns=['required_car_parking_space'])
print("Dropped 'required_car_parking_space' column.")
print(f"Current DataFrame shape: {df.shape}")
```

```python
import os
if not os.path.exists('data'):
    os.makedirs('data')

df.to_csv("Hotel_Reservations_2.csv", index=False)
print("Cleaned data saved to 'Hotel_Reservations_2.csv'.")
```

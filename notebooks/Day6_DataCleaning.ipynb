```python
import pandas as pd
import numpy as np
```

```python
# Load the partially cleaned dataset
df = pd.read_csv('Hotel_Reservations_1.csv')
display(df.head())
```

```python
# Drop the 'Unnamed: 0' column if it exists and is redundant
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
    print("Dropped 'Unnamed: 0' column.")
else:
    print("'Unnamed: 0' column not found or already dropped.")
```

```python
initial_rows = df.shape[0]
invalid_date_rows_to_drop = df[df['arrival_date_full'] == '29-02-2018'].shape[0]
df = df[df['arrival_date_full'] != '29-02-2018']
print(f"Dropped {invalid_date_rows_to_drop} rows with invalid date '29-02-2018'.")
print(f"Remaining rows: {df.shape[0]}")
```

```python
zero_nights_rows_to_drop = df[df['total_nights'] == 0].shape[0]
df = df[df['total_nights'] != 0]
print(f"Dropped {zero_nights_rows_to_drop} rows with zero total nights.")
print(f"Remaining rows: {df.shape[0]}")
```

```python
children_outlier_rows_to_drop = df[df['no_of_children'] > 3].shape[0]
df = df[df['no_of_children'] <= 3]
print(f"Dropped {children_outlier_rows_to_drop} rows with more than 3 children.")
print(f"Remaining rows: {df.shape[0]}")
```

```python
print('Final DataFrame shape:', df.shape)
print('\nFinal DataFrame Info:')
display(df.info())
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Hotel Reservations.csv')
```

```python
df
```

```python
#Feature Enginrring :Handling date and time features
df['arrival_date_full'] = df['arrival_date'].astype(str).str.zfill(2) + '-' + \
                          df['arrival_month'].astype(str).str.zfill(2) + '-' + \
                          df['arrival_year'].astype(str)
```

```python
#Drop the original seprate date columns as they are now combined
df=df.drop(columns=['arrival_year','arrival_month','arrival_date'])
```

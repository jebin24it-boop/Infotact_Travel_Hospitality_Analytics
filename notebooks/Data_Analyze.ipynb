```python
import pandas as pd
```

```python
df=pd.read_csv('Cleaned_Hotel_Reservations.csv')
```

```python
df.dtypes
```

```python
categorical_cols = df.select_dtypes(include='object').columns
print("Categorical Columns:", categorical_cols)
```

```python
print("\n--- Analysis of Categorical Columns ---")
for col in categorical_cols:
    unique_values = df[col].nunique()
    print(f"\nColumn: '{col}'")
    print(f"  Number of unique values: {unique_values}")
    print(f"  Sample values: {df[col].unique()[:5]}") # Display first 5 unique values

    if col in ['Booking_ID', 'arrival_date_full']:
        print("  **Recommendation**: These columns have a very high number of unique values. 'Booking_ID' is likely an identifier and can often be dropped or used for specific grouping/aggregation but usually not for direct encoding. 'arrival_date_full' is a date string and should be converted to datetime objects for feature engineering (e.g., extract year, month, day of week).")
    elif unique_values <= 10: # A common heuristic for one-hot encoding
        print("  **Recommendation**: Given the small number of unique values, **One-Hot Encoding** is often suitable if there's no inherent order (nominal). If there's an order, **Ordinal Encoding** could be considered.")
    else:
        print("  **Recommendation**: For a moderate to high number of unique values, consider **Target Encoding**, **Frequency Encoding**, or **Binary Encoding** to avoid high dimensionality from One-Hot Encoding. If there's a natural order, **Ordinal Encoding** might work.")
```

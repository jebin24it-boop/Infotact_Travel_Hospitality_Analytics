```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Hotel_Reservations.csv')
```

```python
# Define mapping for room types
room_type_mapping = {
    'Room_Type 1': 'Standard Room',
    'Room_Type 4': 'Family Room',
    'Room_Type 2': 'Deluxe Room',
    'Room_Type 6': 'Studio Room',
    'Room_Type 5': 'Executive Room',
    'Room_Type 7': 'Presidential Suite',
    'Room_Type 3': 'Suite'
}

# Define mapping for meal plans
meal_plan_mapping = {
    'Meal Plan 1': 'Breakfast, Lunch & Dinner',
    'Not Selected': 'No Meal Plan',
    'Meal Plan 2': 'Breakfast & Dinner',
    'Meal Plan 3': 'Breakfast Only'
}

# Create new 'descriptive_room_type' column
df['descriptive_room_type'] = df['room_type_reserved'].map(room_type_mapping)

# Create new 'descriptive_meal_plan' column
df['descriptive_meal_plan'] = df['type_of_meal_plan'].map(meal_plan_mapping)

print('New descriptive columns added. Displaying a sample:')
display(df[['room_type_reserved', 'descriptive_room_type', 'type_of_meal_plan', 'descriptive_meal_plan']].head())
```

```python
print('Value counts for descriptive_room_type:')
display(df['descriptive_room_type'].value_counts())

print('\nValue counts for descriptive_meal_plan:')
display(df['descriptive_meal_plan'].value_counts())
```

```python
print('Missing values per column:')
display(df.isnull().sum().sort_values(ascending=False))
```

```python
print('DataFrame Info (Data Types and Non-Null Counts):')
display(df.info())
```

```python
# Convert 'arrival_date_full' to datetime objects
df['arrival_date_full'] = pd.to_datetime(df['arrival_date_full'], format='%d-%m-%Y', errors='coerce')

print("Converted 'arrival_date_full' to datetime.")
```

```python
# Drop the 'Unnamed: 0' column if it exists and is redundant
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
    print("Dropped 'Unnamed: 0' column.")
else:
    print("'Unnamed: 0' column not found.")
```

```python
# Display info again to confirm changes
print('\nDataFrame Info after type conversion and column drop:')
display(df.info())
```

```python
# Identify and count NaT values in 'arrival_date_full'
nat_count = df['arrival_date_full'].isnull().sum()
print(f"Number of NaT values in 'arrival_date_full': {nat_count}")

# Drop rows where 'arrival_date_full' is NaT
df = df.dropna(subset=['arrival_date_full'])
print(f"Dropped {nat_count} rows with NaT values in 'arrival_date_full'.")
```

```python
# Display info again to confirm all 'arrival_date_full' values are non-null
print('\nDataFrame Info after dropping NaT values:')
display(df.info())
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Hotel_Reservations.csv')
```

```python
# Define mapping for room types
room_type_mapping = {
    'Room_Type 1': 'Standard Room',
    'Room_Type 4': 'Family Room',
    'Room_Type 2': 'Deluxe Room',
    'Room_Type 6': 'Studio Room',
    'Room_Type 5': 'Executive Room',
    'Room_Type 7': 'Presidential Suite',
    'Room_Type 3': 'Suite'
}

# Define mapping for meal plans
meal_plan_mapping = {
    'Meal Plan 1': 'Breakfast, Lunch & Dinner',
    'Not Selected': 'No Meal Plan',
    'Meal Plan 2': 'Breakfast & Dinner',
    'Meal Plan 3': 'Breakfast Only'
}

# Create new 'descriptive_room_type' column
df['descriptive_room_type'] = df['room_type_reserved'].map(room_type_mapping)

# Create new 'descriptive_meal_plan' column
df['descriptive_meal_plan'] = df['type_of_meal_plan'].map(meal_plan_mapping)

print('New descriptive columns added. Displaying a sample:')
display(df[['room_type_reserved', 'descriptive_room_type', 'type_of_meal_plan', 'descriptive_meal_plan']].head())
```
